# 3. UPDATE
3.1 Dropping the Rows

```python
# Drop Rows using the na function.
df.dropna(inplace=True)
```

3.2 Dropping the Columns

```python
# Dropping a column using the drop method
df.drop('PetalWidthCm', axis=1)
```

3.3 Creating Calculated Columns

```python
df['Area Given'] = df['SepalLengthCm'] * df['SepalWidthCm']
df.head()
```

3.4 To update an entire column

```python
df['Area Given'] = 100
df.head()
```

3.5 Updating a Single Value

```python
df.iloc[0, -1 ]= 80
df.head()
```

3.6 Condition based using Apply

```python
df['Area Given Binary'] = df['Area Given'].apply(lambda x: 1 if x==80 else 0)
```

```python
df.head()
```
