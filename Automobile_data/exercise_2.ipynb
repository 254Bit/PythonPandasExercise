#Concatenate two data frames using the following conditions

```python
GermanCars = {'make':['Ford','Mercedes','BMV', 'Audi'], 'Price':[2384,171995,135925,71400]}
japaneseCars = {'make':['Toyota','Honda','Nissan','Mitsubishi'], 'Price':[29995,23600,61500,58900]}
```

```python
germanCars=pd.DataFrame.from_dict(GermanCars)
```

```python
germanCars
```

```python
japaneseCars=pd.DataFrame.from_dict(japaneseCars)
japaneseCars
```

```python
carsDF=pd.concat([germanCars,japaneseCars],keys=['Germany','Japan'])
carsDF
```

#Merge two data frames using the following condition

```python
Car_Price = {'make':['Toyota','Honda','BMV','Audi'], 'Price':[29995,23600,61500,58900]}
car_HorsePower = {'make':['Toyota','Honda','BMV','Audi'], 'Horsepower':[141,80,182,160]}
```

```python
carPrice=pd.DataFrame.from_dict(Car_Price)
carPrice
```

```python
carHorsePower=pd.DataFrame.from_dict(car_HorsePower)
carHorsePower
```

```python
carsDF=pd.merge(carPrice, carHorsePower, on ='make')
carsDF
```
