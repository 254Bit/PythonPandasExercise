```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Iris.csv')
```

```python
df.shape
```

```python
df.info()
```

```python
df.columns
```

```python
Species = df['Species'].value_counts().reset_index()
Species
```

```python
plt.figure(figsize=(8,8))
plt.pie(Species['count'], labels=['Iris-setosa','Iris-versicolor','Iris-virginica'], autopct='%1.3f%%', explode=[0,0,0])
plt.legend(loc='upper left')
plt.show()
```

```python
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', data=df, hue='Species')
plt.title('Sepal Width vs Sepal Length')
plt.show()
```

```python
sns.lmplot(
    x='SepalLengthCm',
    y='SepalWidthCm',
    hue='Species',
    palette='bright',
    data=df
)
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
sns.scatterplot (x='PetalLengthCm',y='PetalWidthCm', data=df, hue='Species')
plt.title('Petal Width vs Petal Length')
plt.show()
```

```python
sns.lmplot(
    x='PetalLengthCm',
    y='PetalWidthCm',
    hue='Species',
    palette='bright',
    data=df
)
plt.title('Petal Lenght vs Petal Width')
plt.show()
```
